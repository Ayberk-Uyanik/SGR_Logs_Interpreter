```python
# Global Functions #
def display_well_logs (logs, formations, casings, wellname, parameter):
    if parameter == "res":
        # Constants #
        log_list1 = ["GR", "DT", "RHOB", "NPHI"]
        log_list1_colours = ["brown", "blue", "black", "green"]
        log_labels = ["GR", "DT", "RHOB", "NPHI", "LLD-LLS", "SP", "CALI-BS"]
        casing_list = ["CasingGR", "CasingDT", "CasingRHOB", "CasingNPHI", "CasingLLD", "CasingSP", "CasingCALI"]

        # Figure Generation #
        fig = make_subplots(rows=1, cols=7, subplot_titles=([i for i in log_labels]), shared_yaxes=True)

        # Figure Data Selection #
        for i in range (0, len(log_list1), 1):
            fig.add_trace(go.Scatter(x=logs[f"{log_list1[i]}"], y=logs["DEPTH"], name=f"{wellname} {log_list1[i]} Log", mode="lines", 
                                    line=dict(width=0.5, color=f"{log_list1_colours[i]}"), showlegend=False), row=1, col=i+1)

        fig.add_trace(go.Scatter(x=logs["LLD"], y=logs["DEPTH"], name="{} LLD Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="blue"), showlegend=False), row=1, col=5)
        fig.add_trace(go.Scatter(x=logs["LLS"], y=logs["DEPTH"], name="{} LLS Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="red"), showlegend=False), row=1, col=5)
        fig.add_trace(go.Scatter(x=logs["MSFL"], y=logs["DEPTH"], name="{} MSFL Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="steelblue"), showlegend=False), row=1, col=5)                        
        fig.add_trace(go.Scatter(x=logs["SP"], y=logs["DEPTH"], name="{} SP Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="purple"), showlegend=False), row=1, col=6)
        fig.add_trace(go.Scatter(x=logs["CALI"], y=logs["DEPTH"], name="{} CALIPER Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="black"), showlegend=False), row=1, col=7)
        fig.add_trace(go.Scatter(x=logs["BS"], y=logs["DEPTH"], name="Bad Hole", mode="lines", 
                                line=dict(width=0.8, color="red", dash="dash"), fill="tonexty", fillcolor="yellow", showlegend=True), row=1, col=7)

        # Figure Styles #
        fig.update_layout (title_text="{} Well Logs".format(wellname), width=1500, height=750, showlegend=True)

        # X-Axis Styles #
        fig.update_xaxes(title_text="GR (API)", range=[0, 200], row=1, col=1)
        fig.update_xaxes(title_text="DT (us/ft)", range=[40, 150], row=1, col=2)
        fig.update_xaxes(title_text="RHOB (g/cm3)", range=[1.6, 3.2], row=1, col=3)
        fig.update_xaxes(title_text="NPHI", range=[-0.15, 0.6], row=1, col=4)
        fig.update_xaxes(type="log", title_text="ohmm", range=[-1, 5], row=1, col=5)
        fig.update_xaxes(title_text="SP (mV)", range=[-250, 500], row=1, col=6)
        fig.update_xaxes(title_text="CALIPER (inch)", range=[6, 25], row=1, col=7)

        # Y Axis Styles #
        fig.update_yaxes(title_text="MD (m.)", autorange="reversed", row=1, col=1)
        for i in range (1, 8, 1):
            fig.update_yaxes(autorange="reversed", row=1, col=i)
        
        # Casing Depths #
        for i in range (0, len(casing_list), 1):
            fig.add_trace(go.Scatter(x=casings[f"{casing_list[i]}"], y=casings["DEPTH"], name="Casings", mode="markers+lines", 
                                    line=dict(width=2, color="black"), marker=dict(symbol="triangle-sw", size=18), showlegend=False), row=1, col=i+1) 

    elif parameter == "nores":
        # Constants #
        log_list1 = ["GR", "DT", "RHOB", "NPHI"]
        log_list1_colours = ["brown", "blue", "black", "green"]
        log_labels = ["GR", "DT", "RHOB", "NPHI", "CALI-BS"]
        casing_list = ["CasingGR", "CasingDT", "CasingRHOB", "CasingNPHI", "CasingCALI"]

        # Figure Generation #
        fig = make_subplots(rows=1, cols=5, subplot_titles=([i for i in log_labels]), shared_yaxes=True)

        # Figure Data Selection #
        for i in range (0, len(log_list1), 1):
            fig.add_trace(go.Scatter(x=logs[f"{log_list1[i]}"], y=logs["DEPTH"], name=f"{wellname} {log_list1[i]} Log", mode="lines", 
                                    line=dict(width=0.5, color=f"{log_list1_colours[i]}"), showlegend=False), row=1, col=i+1)

        fig.add_trace(go.Scatter(x=logs["CALI"], y=logs["DEPTH"], name="{} CALIPER Log".format(wellname), mode="lines", 
                                line=dict(width=0.5, color="black"), showlegend=False), row=1, col=5)
        fig.add_trace(go.Scatter(x=logs["BS"], y=logs["DEPTH"], name="Bad Hole", mode="lines", 
                                line=dict(width=0.8, color="red", dash="dash"), fill="tonexty", fillcolor="yellow", showlegend=True), row=1, col=5)

        # Figure Styles #
        fig.update_layout (title_text="{} Well Logs".format(wellname), width=1100, height=750, showlegend=True)

        # X-Axis Styles #
        fig.update_xaxes(title_text="GR (API)", range=[0, 200], row=1, col=1)
        fig.update_xaxes(title_text="DT (us/ft)", range=[40, 150], row=1, col=2)
        fig.update_xaxes(title_text="RHOB (g/cm3)", range=[1.6, 3.3], row=1, col=3)
        fig.update_xaxes(title_text="NPHI", range=[-0.15, 0.6], row=1, col=4)
        fig.update_xaxes(title_text="CALIPER (inch)", range=[6, 25], row=1, col=5)

        # Y Axis Styles #
        fig.update_yaxes(title_text="MD (m.)", autorange="reversed", row=1, col=1)
        for i in range (1, 5, 1):
            fig.update_yaxes(autorange="reversed", row=1, col=i)
        
        # Casing Depths #
        for i in range (0, len(casing_list), 1):
            fig.add_trace(go.Scatter(x=casings[f"{casing_list[i]}"], y=casings["DEPTH"], name="Casings", mode="markers+lines", 
                                    line=dict(width=2, color="black"), marker=dict(symbol="triangle-sw", size=18), showlegend=False), row=1, col=i+1) 

    # Formation Depths #
    for i in range (0, len(formations["Names"]), 1):
      fig.add_hrect(y0=formations["Top"][i], y1=formations["Base"][i], line_width=0, fillcolor=formations["Colours"][i], 
                    opacity=0.2, row=1, col="all", annotation_text=formations["Names"][i], annotation_position="right") 

    fig.show ()
```
