```python
# Calculation of Th/U, Th/K, Depositional Environments and Radiogenic Heat Flows #
# Declaring Functions #
def calculate_proportions (sgr_logs):
    start = 0
    stop = len(sgr_logs["THOR"]) 

    # Proportions #
    sgr_logs["Th/U"] = [sgr_logs["THOR"][i]/sgr_logs["URA"][i] for i in range (start, stop, 1)]
    sgr_logs["Th/K"] = [sgr_logs["THOR"][i]/sgr_logs["POTA"][i] for i in range (start, stop, 1)]
    sgr_logs["U/K"] = [sgr_logs["URA"][i]/sgr_logs["POTA"][i] for i in range (start, stop, 1)]

    return sgr_logs

def calculate_continental_conditions(sgr_logs):
    continental_conditions = sgr_logs[sgr_logs["Th/U"]>7]
    continental_conditions["ENV_DEPTH"] = [(i*0)+7 for i in continental_conditions["Th/U"]]

    return continental_conditions

def calculate_marine_conditions(sgr_logs):    
    marine_conditions = sgr_logs[sgr_logs["Th/U"]<7]
    marine_conditions["ENV_DEPTH"] = [(i*0)+7 for i in marine_conditions["Th/U"]]

    return marine_conditions

def calculate_environments(sgr_logs):
    marine_list = []
    continental_list = []
    reducing_list = []
    oxidising_list = []
    transition_list = []

    for i in sgr_logs["Th/U"]:
        if i > 7:
            continental = i
        elif i <= 7:
            continental = 7
        continental_list.append(continental)

    for i in sgr_logs["Th/U"]:
        if i < 7:
            marine = i
        elif i >= 7:
            marine = 7
        marine_list.append(marine)

    for i in sgr_logs["Th/U"]:
        if i < 2:
            redux = i
        elif i >= 2:
            redux = 2
        reducing_list.append(redux)

    for i in sgr_logs["Th/U"]:
        if i > 7:
            oxidise = i
        elif i <= 7:
            oxidise = 7
        oxidising_list.append(oxidise)

    for i in sgr_logs["Th/U"]:
        if i > 2:
            transition = i
        elif i <= 2:
            transition = 2
        elif i >= 7:
            transition = 7    
        transition_list.append(transition)

    sgr_logs["ENV_DEPTH"] = [(i*0+7) for i in sgr_logs["THOR"]]
    sgr_logs["REDUCE_DEPTH"] = [(i*0+2) for i in sgr_logs["THOR"]]
    sgr_logs["marine"] = marine_list
    sgr_logs["continental"] = continental_list
    sgr_logs["reducing"] = reducing_list
    sgr_logs["oxidising"] = oxidising_list
    sgr_logs["transition"] = transition_list

    return sgr_logs

def calculate_heat_flows(sgr_logs):
    start = 0
    stop = len(sgr_logs["THOR"]) 
    sgr_logs["A"] = [(0.01*sgr_logs["RHOB"][i])*(3.48*sgr_logs["POTA"][i]+9.52*sgr_logs["URA"][i]+2.56*sgr_logs["THOR"][i]) for i in range (start, stop, 1)]

    return sgr_logs
```
